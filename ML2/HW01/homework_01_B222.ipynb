### SVM

* SVM je vhodný model pro binární klasifikaci. Rozhodovací hranice je vytvořena na základě nějaké podmnožiny bodů(tzv. podpůrných vektorů) trenovaci množiny tak, aby co nejvíce bodů bylo správně predikováno. Pokud jsou data lineárně separabilní, existuje nekonečně mnoho řešení, která dokáží přesně rozdělit data na jednotlivé třídy. Pro nalezení rozhodovací hranice je použit princip největšího odstupu, a v případě lineárně neseparabilních dat je také použit regularizační člen.

* **Hyperparametry**:
    * **C** - regularizační parametr, který kontroluje počet chyb, které jsme ochotni tolerovat na trénovací množině. Čím větší je parametr C, tím slabší je regularizace
    * **kernel** - použita jadrová funkce.


* Na ladění hyperparametru a pro zmenšení počtu kodů, implementoval jsem funkci `SVM`.

````python
def SVM(m_Xtrain, m_Xval, m_ytrain, m_yval, tuning_on, kernel_func):
    if(tuning_on == True):
        val_arr = []
        hyperparam_dict = {
            'C': np.arange(0.1, 20, 0.4),
            'kernel':['linear', 'poly', 'rbf', 'sigmoid',]
        }

        param = ParameterGrid(hyperparam_dict)
        for hyp in param:
            clf = SVC(C = hyp['C'], kernel = hyp['kernel'])
            clf.fit(m_Xtrain, m_ytrain)
            val_arr.append(accuracy_score(m_yval, clf.predict(m_Xval)))

        best_params = param[np.argmax(val_arr)]
        print('best params ', best_params)
        tun_clf = SVC(C = best_params['C'], kernel = best_params['kernel'])
        tun_clf.fit(m_Xtrain, m_ytrain)
    else:
        tun_clf = SVC(C = 8.5, kernel = kernel_func)
        tun_clf.fit(m_Xtrain, m_ytrain)

    print('Accuracy score (train): {0:.6f}'.format(accuracy_score(m_ytrain, tun_clf.predict(m_Xtrain))))
    print('Accuracy score (validation): {0:.6f}'.format(accuracy_score(m_yval, tun_clf.predict(m_Xval))))
    print('Support vectors count: ', tun_clf.support_.shape[0])
    print(f'This is {round((tun_clf.support_.shape[0] / m_Xtrain.shape[0])*100, 1)}% of train dataset')
    return accuracy_score(m_yval, tun_clf.predict(m_Xval))
````

* Zavoláme funkci `SVM`.

````python
res_dict["SVM_Tun"] = SVM(Xtrain, Xval, ytrain, yval, True, None)
````

* Z výsledků volání funkce vidíme, že optimální model používá 148 datových bodů, což představuje pouze 10.3% trénovací množině. Používá radiální bazickou funkci a relativně slabou regularizaci. Navíc má docela  dobrou přesnost(0.991667) na validační množině.
* Teď vyzkoušíme jíne jadrové funkce. 

````python
print('Linear kernel')
res_dict["SVM_linK"] = SVM(Xtrain, Xval, ytrain, yval, False, 'linear')
print("------------------------------------")
print('Polynomial kernel')
res_dict["SVM_polyK"] = SVM(Xtrain, Xval, ytrain, yval, False, 'poly')
print("------------------------------------")
print('Sigmoid kernel')
res_dict["SVM_sigmoidK"] = SVM(Xtrain, Xval, ytrain, yval, False, 'sigmoid')
````

### Naivní Bayesův klasifikátor

* **Naivní Bayesův klasifikátor** se dá docela dobře použít na úlohu binární klasifikace. Funguje na základě podmíněných pravděpodobností a díky tomu, že odhad podmíněných pravděpodobností každého příznaku probíhá **nezávisle** na ostatních, je odolný proti problémům s dimenzionalitou. V případě našich dat taková vlastnost může být velmi užitečná.
* V našem případě jsou všechny příznaky numerické, tedy můžeme se pokusit použit naivní bayesův klasifikátor pro normální rozdělení. Předem ale ověříme, zda podmíněna rozděleni $P(X_i | Y = y)$ jsou normální.

* Vytvořil jsem pomocnou funkci `NBclassifier`, která natrénuje model na vstupních datech a vrátí základní informaci o trénování a přesnost na trénováci a validační množině.

````python
def NBclassifier(m_Xtrain, m_Xval, m_ytrain, m_yval):
    clf = GaussianNB()
    clf.fit(m_Xtrain, m_ytrain)
    print('Classes:', clf.classes_)
    print(f"Number of training samples observed in each class: {clf.class_count_}")
    print(f"Probability of each class: {np.round(clf.class_prior_, 3)}")
    print('Accuracy score (train): {0:.6f}'.format(accuracy_score(m_ytrain, clf.predict(m_Xtrain))))
    print('Accuracy score (validation): {0:.6f}'.format(accuracy_score(m_yval, clf.predict(m_Xval))))
    return accuracy_score(m_yval, clf.predict(m_Xval))
````

* Aplikujeme funkci na **ne**modifikována data

````python
res_dict["NB"] = NBclassifier(Xtrain, Xval, ytrain, yval)
````

* Z výpisu funkce vidíme, že počet vzorků v obou třídách je skoro stejný jak i pravděpodobnosti  jednotlivých tříd. Přesnost je už horší v porovnání s SVM ale stejně je docela velká.
* Zkusíme se ještě natrénovat model na standardizovaných a normalizovaných datech.

### LDA

* **LDA** je docela používány generativní model pro klasifikaci(zejména binární). Rozhodovací hranice mezi třídami bude určena nadrovinou. V LDA sdružené podmíněné rozdělení $P(\boldsymbol X | Y = y)$ modelujeme pomocí vícerozměrného normálního rozdělení s dvěma parametry: 
    * ${\boldsymbol\mu}_y$ vektor středních hodnot
    * ${\mathbf \Sigma}$ varianční matice(Pro všechny třídy y stejná)
* Pro případ úlohy binární klasifikaci bude platit:
    
kde 

    

* V průběhů trénování budeme pomoci MLE odhadovat parametry $\hat \pi_y,  \hat{\boldsymbol\mu}_y, \hat{\mathbf \Sigma}$
* Vytvořil jsem pomocnou funkci `LDA`, která natrénuje model na vstupních datech a vrátí základní informaci o trénování a přesnost na trénováci a validační množině.
* Jako `solver`(hyperparametr pro vyhodnocení parametru modelu) používám `svd`. Ten je doporučený `sklearn` pro data s velkým počtem příznaků.

````python
def LDA(m_Xtrain, m_Xval, m_ytrain, m_yval):
    clf = LinearDiscriminantAnalysis(solver = 'svd')
    clf.fit(m_Xtrain, m_ytrain)
    print(f"Probability of each class: {np.round(clf.priors_, 3)}")
    print('Accuracy score (train): {0:.6f}'.format(accuracy_score(m_ytrain, clf.predict(m_Xtrain))))
    print('Accuracy score (validation): {0:.6f}'.format(accuracy_score(m_yval, clf.predict(m_Xval))))
    return accuracy_score(m_yval, clf.predict(m_Xval))
````

* Aplikujeme funkci na **ne**modifikována data

````python
res_dict["LDA"] = LDA(Xtrain, Xval, ytrain, yval)
````

* Z výsledku se dá říct, že pravděpodobnosti jednotlivých tříd jsou skoro stejně jak i v případě naivního bayese. Přesnost je lepší než u naivního bayese ale SVM je stejně lepší.

### PCA a LLE

Zkusíme teď na všechny předchozí modely postupně aplikovat dvě metody redukce dimenzionality. Vzhledem k tomu, že máme velkou dimenzi, redukce dimenzionality může zlepšit přesnost modelů.

* Budeme postupně aplikovat dvě metody:
    * **PCA**(Analýza hlavních komponent) - zkouší najít prostor menší dimenze tak, aby v něm byl co nejlépe zachován celkový rozptyl dat.
    * **LLE**(Lokálně lineární vnoření) - na rozdíl od PCA, LLE hledá prostor menší dimenze takový, aby byla co nejlépe zachována **lokální** struktura(vztahy) dat.

Dále provedeme základní průzkum dat po aplikovaní PCA

* Na vykreslení grafů jsem vytvořil pomocnou funkci `printGraf`, která bude vykreslovat graf na základě vstupních parametrů (pole naměřených hodnot, maximalní hodnota na ose x, názvy os a název grafu). Zelený křížek je maximální hodnota, to se nám bude hodit při zkoumání nejlepšího počtu komponent v LLE a PCA

````python
def printGraf(data_list, max_datapoint, g_name, x_lab, y_lab):
    data_list = np.array(data_list)
    fig, ax = plt.subplots(1,1, figsize=(15, 8))
    ax.scatter(range(1, max_datapoint), data_list, c='red')
    ax.scatter(data_list.argmax(), np.max(data_list), marker='X', c='green')
    plt.title(g_name)
    plt.xlabel(x_lab)
    plt.ylabel(y_lab)
    plt.show()
````

* Teď zkusíme změřit podíl vysvětleného rozptylu prvními n komponentami. Tato informace nám umožní na začátku zjistit, kolik minimálně komponent budeme potřebovat, abychom zajistili 100% vysvětlení rozptylu.

````python
var_arr_pca = []
max_comp_count = Xtrain.shape[1]
for i in range(1, (max_comp_count+1)):
    pca = PCA(n_components = i)
    Xtrain_pca = pca.fit_transform(Xtrain)
    var_arr_pca.append(round(np.sum(pca.explained_variance_ratio_), 3))
````

````python
printGraf(var_arr_pca, max_comp_count+1, "Proportion of variance explained by the first n components using PCA", 
                                          "components count", "explained var ratio")
````

* Z grafu lze zjistit, že od určitého počtu komponent kolem 500 se podíl vysvětleného rozptylu chová konstantně. Zjistíme přesně, od kolika komponent bude podíl vysvětleného rozptylu 100%.

````python
var_arr_pca = np.array(var_arr_pca)
result = np.where(var_arr_pca == 1)
result
````

* Takže 100% rozptylu v datech je vysvětleno prvními 503 komponentami(v pole zapisujeme 1. komponentu na pozici 0).

* Taky zjistíme rozptyl vysvětlený $i.$ hlavní komponentou.

````python
pca_all = PCA(n_components = 784)
Xtrain_pca_all = pca_all.fit_transform(Xtrain)
np.set_printoptions(suppress = True)
var_arr_comp = np.round(pca_all.explained_variance_, 2)
printGraf(var_arr_comp, max_comp_count+1, "The amount of variance explained by each of the selected component", 
                                          "component number", "explained var ratio")
````

* Graf je docela nepřehledný. Stalo se to kvůli velkému rozsahu hodnot. Maximum je až milion a minimum je nula.

````python
print(f'Maximum: {max(var_arr_comp)}')
print(f'Minimum: {min(var_arr_comp)}')
````

* Z grafu je docela špatně vidět jaké komponenty mají vysvětlený rozptyl nulový proto musíme to zjistit. 

````python
var_arr_comp = np.array(var_arr_comp)
result_all = np.where(var_arr_comp == 0)
result_all
````

* Tedy od 703 komponenty dal všechny komponenty mají 0 vysvětlený rozptyl, proto můžeme jich dále neuvažovat.

* Napsal jsem také pomocnou funkci `transform_and_train`, která na základě vstupních parametrů zredukuje data pomocí metody PCA nebo LLE a aplikuje na ně jeden ze tří modelů, které byly zaznamenány dříve. Funkce najde optimální počet komponent a vykreslí graf závislosti přesnosti modelu na počtu komponent. Hyperparametry pro modely použijeme již dříve zjištěné během ladění. 
* Parametry funkce jsou trénovací a validační data, model který chceme natrénovat(`model`), metoda redukce dimenzionality(`transf`), použití normalizovaných/standardizovaných dat(`norm_stand`), pokud používáme nemodifikována data  `None`, jádrová funkce pro SVM(`ker`), pro jiné modely `None`.

````python
def transform_and_train(m_Xtrain, m_Xval, m_ytrain, m_yval, model, transf, norm_stand, ker):
    val_arr = []
    max_comp_count = 702
    for i in range(1, (max_comp_count+1)):
        if(transf == "PCA"):
            mod_transf = PCA(n_components=i)
        elif(transf == "LLE"):
            mod_transf = LocallyLinearEmbedding(n_components = i)
        else:
            print("Error name")
            return
        Xtrain_transf = mod_transf.fit_transform(m_Xtrain)
        Xval_transf = mod_transf.transform(m_Xval)
        
        if(model == "LDA"):
            clf = LinearDiscriminantAnalysis()
        elif(model == "SVM"):
            if(norm_stand == "Normalization"): 
                clf = SVC(C = 2.5, kernel = ker)
            elif(norm_stand == "Standardisation"):
                clf = SVC(C = 0.5, kernel = ker)
            else:
                clf = SVC(C = 8.5, kernel = ker)
        elif(model == "NBClassifier"):
            clf = GaussianNB()
        else:
            print("Error name")
            return
        clf.fit(Xtrain_transf, m_ytrain)
        val_arr.append(accuracy_score(m_yval, clf.predict(Xval_transf)))
    print(f"Best accuracy score is {max(val_arr)} for dimension {np.argmax(val_arr) + 1}")
    if(norm_stand != None):
        printGraf(val_arr, max_comp_count+1, f"{transf} with {model} and {norm_stand}", "components count", "accuracy_score")
    else:
        printGraf(val_arr, max_comp_count+1, f"{transf} with {model}", "components count", "accuracy_score")

    return max(val_arr)
````

Pro velký počet komponent přesnost modelů začíná klesat. Takový skok může být docela podezřelý. To může souviset s tím, že máme úplně negausovská data a projekce do menšího prostoru nám pomůže.
