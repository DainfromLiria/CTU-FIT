````python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import seaborn           as sns

from   sklearn.metrics           import silhouette_score
from   sklearn.cluster           import AgglomerativeClustering
from   scipy.cluster.hierarchy   import dendrogram
from   sklearn.model_selection   import train_test_split
from   sklearn.model_selection   import ParameterGrid
````

### Data:
#### ATRIBUTY*  Customer ID: jedinečné číslo zákazník
* Gender: muž/ žena
* Age: věk zákazíka
* Annual Income(k$): roční příjem zákaz
* Spending Score: interní skóre přiřazené zákazníkovi podle chování a objemu utracených finančních prostředkůředků

````python
data = pd.read_csv('data_customers.csv')
data.head()
````

## Příprava dat

````python
data.info()
````

Příznak `CustomerID` smažeme, protože pro shlukování ho nepotřebujeme.

````python
data = data.drop(columns=['CustomerID'])
data.head()
````

Převedeme `Gender` na category a pak pomocí `cat.codes` na číselnou hodnotu.

````python
data[["Gender"]] = data[["Gender"]].astype('category').apply(lambda x: x.cat.codes)
data.head()
````

Teď **Male** = 1 a **Female** = 0.

Rozdělime data na  trénovací a testovací.
* 70% trénovací
* 30% testovací

````python
random_seed = 33 #constant for random_state

train, test = train_test_split(data, test_size = 0.3, random_state = random_seed) 
````

## Aglomerativní hierarchické shlukování

Zkusíme použit aglomerativní hierarchické shlukování a zobrazit dendrogram, aby zjistit vhodný počet shluků. Nastavíme `distance_threshold`  na nulu, ten řídí maximální vzdálenost pro spojení dvou bodu do shluku. Počet shluku(`n_clusters`) v takovém případě musí být `None`(podle [dokumentace](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html#sklearn.cluster.AgglomerativeClustering.get_metadata_routing)).

````python
aggl_clust = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
aggl_clust = aggl_clust.fit(train)
````

* Aby vytvořit dendrogram, budeme potřebovat pomocnou funkci, která vytvoří matici propojení. Funkce ```plot_dendrogram``` je už implementována v [dokumentaci sklearn](https://scikit-learn.org/stable/auto_examples/cluster/plot_agglomerative_dendrogram.html#sphx-glr-auto-examples-cluster-plot-agglomerative-dendrogram-py). Změníme jenom návratovou hodnotu na matici propojení.

````python
def plot_dendrogram(model):

    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    return linkage_matrix
````

* Nalezneme matici propojení a vytvoříme dendrogram. Použitím parametru ```truncate_mode```, který nastavíme na **lastp**, vykreslíme jenom posledních **p** spojeních shluků. Parametr **p** nastavíme na 25, aby graf byl čitelný. Otočím popisy x osy na 90 stupňů a nastavíme velikost fontu popisů na 12.

````python
mat_link = plot_dendrogram(aggl_clust)
dendrogram(mat_link, truncate_mode="lastp", p=25, leaf_rotation=90, leaf_font_size=12)
plt.show()
````

Z dendrogramu můžeme říct, že vhodný počet shluků bude nejspíš 5.(rozřízneme dendrogram na vzdálenosti 150).

Zjistíme silhouette score pro 5 shluků.

````python
aggl_clust = AgglomerativeClustering(n_clusters=5)
aggl_clust = aggl_clust.fit(train)

print('AgglomerativeClustering Silhouette score: ' + str(silhouette_score(train, aggl_clust.labels_)))
````

* Silhouette score je docela dobrý ale pokusíme se odladit hyperparametry. 
Těch bude dva:
    1. **metric** - metrika
    2. **linkage** - metoda měření vzdálenosti shluků

````python
val_arr_agl = []
hyperparam_arr_agl = {
    'metric': ['euclidean', 'manhattan', 'l1', 'l2'],
    'linkage': ['complete', 'average', 'single'],
}
param_agl = ParameterGrid(hyperparam_arr_agl)
for hyp in param_agl:
    aggl_clust = AgglomerativeClustering(n_clusters = 5, metric = hyp['metric'], linkage = hyp['linkage'], ).fit(train)
    val_arr_agl.append(silhouette_score(train, aggl_clust.labels_))
````

Zjistíme nejlepší hyperparametry(nejvyšší silhouette score).

````python
best_params_agl = param_agl[np.argmax(val_arr_agl)]
print('best params ', best_params_agl)
````

## Metoda pro výpočet vzdálenosti shluků

Po ladění hyperparametrů můžeme jistě říct, že nejlepší je párová vzdálenost (average linkage), což je kompromis mezi metodou nejbližšího souseda a metodou nejvzdálenějšího souseda. Výpočet pro shluky $A$ a $B$ vypadá následovně:

$$D_{\text{avg}}(A, B) = \frac{1}{|A| \times |B|} \sum_{\substack{x \in A \\ y \in B}} d(x, y)$$, kde $d(x, y)$ je vzdálenost mezi body $x$ a $y$, v našem případě euklidovská.


Provedeme shlukování s požitím těchto hodnot hyperparametrů.

````python
aggl_clust = AgglomerativeClustering(n_clusters = 5, metric = best_params_agl['metric'], linkage = best_params_agl['linkage']).fit(train)
print('AgglomerativeClustering Silhouette score: ' + str(silhouette_score(train, aggl_clust.labels_)))
````


Přidejme číslo shluku jako sloupec do našeho datasetu.

````python
train["cluster"] = aggl_clust.labels_
train
````


Taky zjistíme počet bodů v každém ze shluků.

````python
unique, counts = np.unique(aggl_clust.labels_, return_counts=True)
fin_clusters = dict(zip(unique, counts))
fin_clusters
````

Z uvedených pozorování sestavíme pravidla.

**Pravidla**:
* **1**. **IF** annual_income > 70 AND annual_income < 137 AND spending_score > 1 AND spending_score < 40  **THEN** cluster = 0
* **2**. **IF** annual_income > 38 AND annual_income < 69 AND spending_score > 35 AND spending_score < 65  **THEN** cluster = 1
* **3**. **IF** annual_income > 15 AND annual_income < 34 AND spending_score > 4 AND spending_score < 40  **THEN** cluster = 2
* **4**. **IF** annual_income > 15 AND annual_income < 38 AND spending_score > 72 AND spending_score < 99 AND age < 40 **THEN** cluster = 3
* **5**. **IF** annual_income > 69 AND annual_income < 103 AND spending_score > 63 AND spending_score < 97 AND age < 40 **THEN** cluster = 4

## Implementace pravidel

Vytvoříme funkci `rules`, ve které implementujeme všechna výše uvedená pravidla.

````python
def rules(annual_income, spending_score, age):
    if(annual_income > 70 and annual_income < 137) and (spending_score > 1 and spending_score < 40):
        return 0
    if(annual_income > 38 and annual_income < 69) and (spending_score > 35 and spending_score < 65):
        return 1
    if(annual_income > 15 and annual_income < 34) and (spending_score > 4 and spending_score < 40):
        return 2
    if(annual_income > 15 and annual_income < 38) and (spending_score > 72 and spending_score < 99) and age < 40:
        return 3
    if(annual_income > 69 and annual_income < 103) and (spending_score > 63 and spending_score < 97) and age < 40:
        return 4
    return -1
````

## Testování kvality pravidel

Pomocí funkce `rules` napočítáme predikcí pravidel na testovacích datech.

````python
rule_predict = []
for index, row in test.iterrows():
    rule_predict.append(rules(row['Annual Income (k$)'], row['Spending Score (1-100)'], row['Age']))
````

Teď zjistíme správné shluky pro jednotlivé body testovací množiny. Každý nový bod z testovací množiny bude nový shluk a pomocí euklidovské vzdálenosti pro body a párové vzdáleností pro shluky určíme, ke kterému shluku tento bod přiřadíme.

````python
real_cluster = []
for index, row in test.iterrows():
    tmp = []
    for i in range(0, 5):
        sum = 0
        for index, row2 in train[train["cluster"] == i].iterrows():
            sum += np.sqrt(  (row2["Age"] - row["Age"])**2
                    + (row2["Annual Income (k$)"] - row["Annual Income (k$)"])**2 
                    + (row2["Spending Score (1-100)"] - row["Spending Score (1-100)"])**2 )
        tmp.append((1/len(train[train["cluster"] == i]))*sum)
    real_cluster.append(np.argmin(tmp))
````

Zjistíme přesnost jako počet správně predikovaných bodů dělený celkovým počtem bodů.

````python
correct = 0
for i in range(0, len(real_cluster)):
    if rule_predict[i] == real_cluster[i]:
        correct += 1
print(f"Prediction score is {correct / len(real_cluster)}")
print("Correct: ", correct)
print("Wrong: ", len(real_cluster) - correct)
````

Výsledek je docela dobrý, 78% správně predikovaných bodů na testovacích datech.
