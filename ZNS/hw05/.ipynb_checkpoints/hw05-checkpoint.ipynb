Z histogramů lze vyčíst, že příznaky `PetalWidthCm`, `PetalLengthCm` a `SepalLengthCm` budou mít 3 možné hodnoty lingvistických proměnných, zatímco `SepalWidthCm` bude mít jen jednu (histogram naznačuje, že to bude normální rozdělení). Tyto 4 proměnné budou našimi vstupními lingvistickými proměnnými. Výstupní proměnnou bude `Species`, která bude mít jako hodnuty názvy jednotlivých druhů kosatců. Jako funkci příslušnosti použijeme trojúhelníkovou funkci, která je již implementována v balíčku `fuzzy_expert` jako `trimf`.

```python
variables = {
    "PetalWidthCm": FuzzyVariable(
        universe_range=(0, 2.6),
        terms={
            "Long": ('trimf', 1.6, 2, 2.5),
            "Middle": ("trimf", 0.6, 1.5, 1.6),
            "Short": ('trimf', 0.1, 0.1, 0.6),
        },
    ),
    
    "PetalLengthCm": FuzzyVariable(
        universe_range=(0.9, 7),
        terms={
            "Long": ('trimf', 5, 6, 7),
            "Middle": ("trimf", 3, 4, 5),
            "Short": ('trimf', 1, 1.5, 3),
        },
    ),
    
    "SepalWidthCm": FuzzyVariable(
        universe_range=(1.9, 4.5),
        terms={
            "Const": ('trimf', 2, 3.4, 4.4),
        },
    ),

    "SepalLengthCm": FuzzyVariable(
        universe_range=(4, 8),
        terms={
            "Long": ('trimf', 6, 6.5, 8),
            "Middle": ('trimf', 5.3, 5.7, 6),
            "Short": ('trimf', 4.5, 5, 5.5),
        },
    ),

    "Species": FuzzyVariable(
        universe_range=(0, 2),
        terms={
            "Iris-virginica": [(1,0), (2,1)], # Triangular membership function.
            "Iris-versicolour": [(0,0), (1,1), (2,0)], # Triangular membership function.
            "Iris-setosa": [(0,1), (1,0)], # Triangular membership function.
        },
    ),
}
```

## Fuzzy pravidla

Použil jsem již existující pravidla z [cvičeni 9](https://courses.fit.cvut.cz/BI-ZNS/tutorials/media_tutorials_2021_web_cvic09.html). Přepsal jsem je do formátu fuzzy pravidel pro další práci.

```python
rules = [
 FuzzyRule(
      premise=[
            ("PetalWidthCm", "Short")
     ],
     consequence=[("Species", "Iris-setosa")],
 ),
  FuzzyRule(
      premise=[
            ("PetalWidthCm", "Long"),
        ],
        consequence=[("Species", "Iris-virginica")],
  ),
   FuzzyRule(
      premise=[
            ("PetalLengthCm", "Middle"),
            ("AND", "PetalWidthCm", "Middle"),
        ],
        consequence=[("Species", "Iris-versicolour")],
  ),
   FuzzyRule(
      premise=[
            ("PetalLengthCm", "Long"),
            ("AND", "PetalWidthCm", "Middle"),
        ],
        consequence=[("Species", "Iris-virginica")],
  ),
]
```

## Defuzzyfikace

Použil jsem inferenční mechanismus implementovaný v balíčku `fuzzy_expert`. Zkusil jsem 3 metody defuzifikace:
1. `cog` - těžiště.
2. `boa` - bisekce oblasti.
3. `mom` - průměr hodnot pro které funkce přislušnosti nabivá maximum.

```python
model_cog = DecompositionalInference(
    and_operator="min",
    or_operator="max",
    implication_operator="Rc",
    composition_operator="max-min",
    production_link="max",
    defuzzification_operator="cog",
)
```

```python
model_boa = DecompositionalInference(
    and_operator="min",
    or_operator="max",
    implication_operator="Rc",
    composition_operator="max-min",
    production_link="max",
    defuzzification_operator="boa",
)
```

```python
model_mom = DecompositionalInference(
    and_operator="min",
    or_operator="max",
    implication_operator="Rc",
    composition_operator="max-min",
    production_link="max",
    defuzzification_operator="mom",
)
```

Funkce `apply_rules` pro vstupní inferenční mechanismus vypočítá predikci na testovací množině a vrátí pole predikcí pro každý bod trénovací množiny.

```python
def apply_rules(model):
    predicted = []
    for index, row in Xtest.iterrows():
        tmp = model(
          variables=variables,
          rules=rules,
          SepalLengthCm=row["SepalLengthCm"],
          SepalWidthCm=row["SepalWidthCm"],
          PetalLengthCm=row["PetalLengthCm"],
          PetalWidthCm=row["PetalWidthCm"],
         )
        pred = round(tmp[0].get("Species"), 5)
        #print(pred)
        if pred == 1.0:
            predicted.append(1)
        elif pred < 1.0:
            predicted.append(0)
        elif pred > 1.0:
            predicted.append(2)
    return predicted
```

Funkce `Accuracy` spočítá, kolik bodů bylo klasifikováno správně, a vypíše přesnost, počet správně a špatně klasifikovaných bodů na trénovací množině.

```python
def Accuracy(predicted_data):
    i = 0
    correct = 0
    df_len = Xtest["Species"].count()
    for index, row in Xtest.iterrows():
       if row["Species"] == predicted_data[i]:
           correct += 1
       i += 1

    print(f"Prediction score is {correct / df_len}")
    print("Correct: ", correct)
    print("Wrong: ", df_len - correct)
```

Aplikujeme předchozí funkce na jednotlivé varianty inferenčního mechanismu.

```python
pred_cog = apply_rules(model_cog)
pred_boa = apply_rules(model_boa)
pred_mom = apply_rules(model_mom)
```

```python
print("Center of gravity: \n")
Accuracy(pred_cog)
print("====================")
print("Bisector of area: \n")
Accuracy(pred_boa)
print("====================")
print("Mean of the values for which the membership function is maximum: \n")
Accuracy(pred_mom)
```

Z výsledků funkce Accuracy vidíme, že defuzzyfikace metodou těžiště a bisekce oblasti dává stejný výsledek a je nejlepší. Metoda průměr maxim nesprávně klasifikovala jen o jeden záznam navíc. Jako výsledek můžeme zvolit libovolnou z těchto dvou metod, tedy těžiště nebo bisekci oblasti..
